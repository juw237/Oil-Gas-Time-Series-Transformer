```python
trues_over=[]
preds_over=[]
term=df1.shape[0]
for i in range(term):
    trues_over.append(df2[i,1,-1])
    preds_over.append(df1[i, 1, -1])

import matplotlib.pyplot as plt
import seaborn as sns
# draw OT prediction

plt.figure(figsize=(10,5))
plt.plot(trues_over, c='red',label='GroundTruth')
plt.plot(preds_over,c='purple', label='Prediction ns Informer')
plt.legend(fontsize=15)
plt.title('One Step Surfase Torque Prediction Vs Ground Truth on Test dataset', fontsize=15)
plt.xlabel('Time Sequence', fontsize=15)
plt.ylabel('Predictions', fontsize=15)
plt.savefig('555.png',dpi=100)
plt.show()
```
